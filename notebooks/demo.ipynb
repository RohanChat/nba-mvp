```python
game_id = '0029600012'

query = f"""
WITH player1_stats AS (
    SELECT game_id, player1_id AS player_id, homedescription, neutraldescription, visitordescription FROM play_by_play
    WHERE game_id = '{game_id}'
    AND player1_id IS NOT NULL
), player2_stats AS (
    SELECT game_id, player2_id AS player_id, homedescription, neutraldescription, visitordescription FROM play_by_play
    WHERE game_id = '{game_id}'
    AND player2_id IS NOT NULL
), player3_stats AS (
    SELECT game_id, player3_id AS player_id, homedescription, neutraldescription, visitordescription FROM play_by_play
    WHERE game_id = '{game_id}'
    AND player3_id IS NOT NULL
), all_players AS (
    SELECT * FROM player1_stats
    UNION ALL
    SELECT * FROM player2_stats
    UNION ALL
    SELECT * FROM player3_stats
)

SELECT 
    player_id,
    homedescription,
    neutraldescription,
    visitordescription
FROM all_players
"""

df = pd.read_sql(query, conn)
print(f"-------READING FROM -- {table} TABLE-------")
for row in df.itertuples():
    print(row)
```
