```python
import numpy as np

# split the play by play for this game id into chunks of roughly 4 minutes, by utilising the pctimestring column 6

def get_play_by_play_in_chunks(game_id, chunk_minutes=4):
    """
    Retrieve play-by-play data for a specific game and split it into chunks of specified minutes.
    
    Parameters:
    - game_id: The ID of the game to retrieve data for
    - chunk_minutes: Size of each chunk in minutes (default: 4)
    
    Returns:
    - DataFrame with play-by-play data and an additional 'time_chunk' column
    """

    
    # Retrieve play-by-play data for the specified game
    query = f"""
    SELECT *
    FROM play_by_play
    WHERE game_id = '{game_id}'
    ORDER BY period, eventnum
    """
    
    df = pd.read_sql_query(query, conn)
    
    if df.empty:
        return "No play-by-play data found for this game."
    
    # Function to convert pctimestring to seconds remaining in the quarter
    def time_to_seconds(time_str):
        if pd.isna(time_str):
            return np.nan
        parts = time_str.split(':')
        if len(parts) == 2:
            minutes, seconds = parts
            return int(minutes) * 60 + float(seconds)
        return np.nan
    
    # Convert pctimestring to seconds remaining in quarter
    df['seconds_remaining'] = df['pctimestring'].apply(time_to_seconds)
    
    # Calculate game time in seconds (0 = start of game)
    df['game_seconds'] = (df['period'] - 1) * 12 * 60 + (12 * 60 - df['seconds_remaining'])
    
    # Determine which chunk each play belongs to
    chunk_seconds = chunk_minutes * 60
    df['time_chunk'] = (df['game_seconds'] // chunk_seconds) + 1
    
    # Create a descriptive label for each chunk
    def format_chunk_label(chunk_num, chunk_minutes):
        start_min = (chunk_num - 1) * chunk_minutes
        end_min = chunk_num * chunk_minutes
        start_qtr = start_min // 12 + 1
        start_qtr_min = start_min % 12
        end_qtr = end_min // 12 + 1
        end_qtr_min = end_min % 12
        
        if start_qtr == end_qtr:
            return f"Q{start_qtr}: {start_qtr_min}-{end_qtr_min} min"
        else:
            return f"Q{start_qtr}: {start_qtr_min} min - Q{end_qtr}: {end_qtr_min} min"
    
    df['chunk_label'] = df['time_chunk'].apply(lambda x: format_chunk_label(x, chunk_minutes))
    
    return df

# Example usage
game_id = "0042100201"  # Replace with an actual game ID
chunked_pbp = get_play_by_play_in_chunks(game_id)

# To see the chunks distribution
if isinstance(chunked_pbp, pd.DataFrame):
    chunk_summary = chunked_pbp.groupby(['time_chunk', 'chunk_label']).size().reset_index(name='play_count')
    print(chunk_summary)

# print the first 5 and last 5 tows in the chunked play by play
display(chunked_pbp.head())
display(chunked_pbp.tail())
```

```python
def add_scores_to_pbp(pbp_df):
    """
    Add home and away scores to play-by-play data.
    
    Parameters:
    - pbp_df: DataFrame containing play-by-play data
    
    Returns:
    - DataFrame with additional 'pts_home' and 'pts_away' columns
    """
    
    # Retrieve line score data for the specified game
    game_id = pbp_df['game_id'].iloc[0]
    query = f"""
    SELECT *
    FROM line_score
    WHERE game_id = '{game_id}'
    """
    
    line_score = pd.read_sql_query(query, conn)
    
    if line_score.empty:
        return "No line score data found for this game."
    
    # Get the scalar values from the line_score DataFrame
    home_score = line_score['pts_home'].iloc[0]
    away_score = line_score['pts_away'].iloc[0]
    
    # Add these scores to every row in the play-by-play DataFrame
    pbp_df = pbp_df.copy()  # Create a copy to avoid SettingWithCopyWarning
    pbp_df['pts_home'] = home_score
    pbp_df['pts_away'] = away_score
    
    return pbp_df

# Example usage
chunked_pbp_scores = add_scores_to_pbp(chunked_pbp)

# print the first 5 and last 5 rows in the chunked play by play
display(chunked_pbp_scores.head())
display(chunked_pbp_scores.tail())
```

```python
import pandas as pd
import numpy as np

def get_play_by_play_in_chunks_csv(game_id, csv_file, chunk_minutes=4):
    """
    Retrieve play-by-play data for a specific game from a CSV file (with columns like 'gameid', 'period', 'clock', etc.)
    and split it into chunks of specified minutes.
    
    Parameters:
    - game_id: The ID of the game to retrieve data for.
    - csv_file: Path to the CSV file containing play-by-play data.
    - chunk_minutes: Size of each chunk in minutes (default: 4).
    
    Returns:
    - DataFrame with play-by-play data for the game and additional columns:
      'seconds_remaining', 'game_seconds', 'time_chunk', and 'chunk_label'.
    """
    # Read the CSV file
    df = pd.read_csv(csv_file)
    
    # Determine the correct game ID column name
    if 'gameid' in df.columns:
        game_col = 'gameid'
    else:
        return "No game id column found in CSV."
    
    # Filter for the specific game
    df_game = df[df[game_col] == game_id].copy()
    print(df[game_col])
    if df_game.empty:
        return "No play-by-play data found for this game."
    
    # Parse the 'clock' column, which is in an ISO 8601 duration format (e.g., "PT12M00.00S")
    def parse_clock(clock_str):
        if pd.isna(clock_str):
            return np.nan
        clock_str = clock_str.strip()
        if not clock_str.startswith("PT"):
            return np.nan
        # Remove the "PT" prefix.
        clock_str = clock_str[2:]
        # Look for the minutes separator.
        if "M" in clock_str:
            parts = clock_str.split("M")
            try:
                minutes = int(parts[0])
            except:
                minutes = 0
            seconds_str = parts[1].rstrip("S")
            try:
                seconds = float(seconds_str) if seconds_str != "" else 0.0
            except:
                seconds = 0.0
            return minutes * 60 + seconds
        else:
            # If there is no 'M', assume the string is just seconds.
            seconds_str = clock_str.rstrip("S")
            try:
                return float(seconds_str)
            except:
                return np.nan

    # Create a new column for seconds remaining in the period
    df_game['seconds_remaining'] = df_game['clock'].apply(parse_clock)
    
    # Assume each period is 12 minutes (720 seconds). Compute the elapsed game seconds.
    period_length = 720
    # Here, we treat the game time as: (period - 1) * period_length + (period_length - seconds_remaining)
    # so that at the start of a period (clock = PT12M00.00S) game_seconds is (period-1)*720.
    df_game['game_seconds'] = (df_game['period'] - 1) * period_length + (period_length - df_game['seconds_remaining'])
    
    # Determine which chunk each play belongs to.
    chunk_seconds = chunk_minutes * 60
    df_game['time_chunk'] = (df_game['game_seconds'] // chunk_seconds) + 1
    
    # Generate a descriptive label for each chunk.
    def format_chunk_label(chunk_num, chunk_minutes):
        start_min = (chunk_num - 1) * chunk_minutes
        end_min = chunk_num * chunk_minutes
        start_qtr = start_min // 12 + 1
        start_qtr_min = start_min % 12
        end_qtr = end_min // 12 + 1
        end_qtr_min = end_min % 12
        if start_qtr == end_qtr:
            return f"Q{start_qtr}: {start_qtr_min}-{end_qtr_min} min"
        else:
            return f"Q{start_qtr}: {start_qtr_min} min - Q{end_qtr}: {end_qtr_min} min"
    
    df_game['chunk_label'] = df_game['time_chunk'].apply(lambda x: format_chunk_label(x, chunk_minutes))
    
    return df_game

# Example usage:
csv_file = pd.read_csv("../../data/nba_csv/pbp/pbp2022.csv")
game_id = 52100111 
new_chunked_pbp = get_play_by_play_in_chunks_csv(game_id, "../../data/nba_csv/pbp/pbp2022.csv")

print(new_chunked_pbp)
```

```python
# for all uniqur game ids, chunk the data and then merge the data for that game id with the entries in ../../data/nba_csv/traditional.csv
# we want to add all columns from the entry in traditional.csv to every entry in the chunked play by play data for that game id, except for 'type' and 'gameid'

def merge_pbp_with_traditional(game_id, pbp_df, traditional_file):
    """
    Merge play-by-play data with traditional data for a specific game.
    
    Parameters:
    - game_id: The ID of the game to merge data for.
    - pbp_df: DataFrame containing play-by-play data.
    - traditional_file: Path to the CSV file containing traditional data.   "
    """
    # Read the traditional data
    df_traditional = pd.read_csv(traditional_file)
    
    # Filter for the specific game
    df_traditional_game = df_traditional[df_traditional['gameid'] == game_id].copy()
    
    # Check if we have data for this game
    if df_traditional_game.empty:
        print(f"No traditional data found for game {game_id}")
        return
        
    # Merge using a cartesian product approach
    df_merged = pd.merge(pbp_df, df_traditional_game, on='team', how='left')

    return df_merged

# Example usage:
traditional_file = "../../data/nba_csv/team_traditional.csv"
merged_data = merge_pbp_with_traditional(game_id, new_chunked_pbp, traditional_file)
print(merged_data)
```
