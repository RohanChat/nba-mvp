```python
import numpy as np

# split the play by play for this game id into chunks of roughly 4 minutes, by utilising the pctimestring column 6

def get_play_by_play_in_chunks(game_id, chunk_minutes=4):
    """
    Retrieve play-by-play data for a specific game and split it into chunks of specified minutes.
    
    Parameters:
    - game_id: The ID of the game to retrieve data for
    - chunk_minutes: Size of each chunk in minutes (default: 4)
    
    Returns:
    - DataFrame with play-by-play data and an additional 'time_chunk' column
    """

    
    # Retrieve play-by-play data for the specified game
    query = f"""
    SELECT *
    FROM play_by_play
    WHERE game_id = '{game_id}'
    ORDER BY period, eventnum
    """
    
    df = pd.read_sql_query(query, conn)
    
    if df.empty:
        return "No play-by-play data found for this game."
    
    # Function to convert pctimestring to seconds remaining in the quarter
    def time_to_seconds(time_str):
        if pd.isna(time_str):
            return np.nan
        parts = time_str.split(':')
        if len(parts) == 2:
            minutes, seconds = parts
            return int(minutes) * 60 + float(seconds)
        return np.nan
    
    # Convert pctimestring to seconds remaining in quarter
    df['seconds_remaining'] = df['pctimestring'].apply(time_to_seconds)
    
    # Calculate game time in seconds (0 = start of game)
    df['game_seconds'] = (df['period'] - 1) * 12 * 60 + (12 * 60 - df['seconds_remaining'])
    
    # Determine which chunk each play belongs to
    chunk_seconds = chunk_minutes * 60
    df['time_chunk'] = (df['game_seconds'] // chunk_seconds) + 1
    
    # Create a descriptive label for each chunk
    def format_chunk_label(chunk_num, chunk_minutes):
        start_min = (chunk_num - 1) * chunk_minutes
        end_min = chunk_num * chunk_minutes
        start_qtr = start_min // 12 + 1
        start_qtr_min = start_min % 12
        end_qtr = end_min // 12 + 1
        end_qtr_min = end_min % 12
        
        if start_qtr == end_qtr:
            return f"Q{start_qtr}: {start_qtr_min}-{end_qtr_min} min"
        else:
            return f"Q{start_qtr}: {start_qtr_min} min - Q{end_qtr}: {end_qtr_min} min"
    
    df['chunk_label'] = df['time_chunk'].apply(lambda x: format_chunk_label(x, chunk_minutes))
    
    return df

# Example usage
game_id = "0042100201"  # Replace with an actual game ID
chunked_pbp = get_play_by_play_in_chunks(game_id)

# To see the chunks distribution
if isinstance(chunked_pbp, pd.DataFrame):
    chunk_summary = chunked_pbp.groupby(['time_chunk', 'chunk_label']).size().reset_index(name='play_count')
    print(chunk_summary)

# print the first 5 and last 5 tows in the chunked play by play
display(chunked_pbp.head())
display(chunked_pbp.tail())
```

```python
def add_scores_to_pbp(pbp_df):
    """
    Add home and away scores to play-by-play data.
    
    Parameters:
    - pbp_df: DataFrame containing play-by-play data
    
    Returns:
    - DataFrame with additional 'pts_home' and 'pts_away' columns
    """
    
    # Retrieve line score data for the specified game
    game_id = pbp_df['game_id'].iloc[0]
    query = f"""
    SELECT *
    FROM line_score
    WHERE game_id = '{game_id}'
    """
    
    line_score = pd.read_sql_query(query, conn)
    
    if line_score.empty:
        return "No line score data found for this game."
    
    # Get the scalar values from the line_score DataFrame
    home_score = line_score['pts_home'].iloc[0]
    away_score = line_score['pts_away'].iloc[0]
    
    # Add these scores to every row in the play-by-play DataFrame
    pbp_df = pbp_df.copy()  # Create a copy to avoid SettingWithCopyWarning
    pbp_df['pts_home'] = home_score
    pbp_df['pts_away'] = away_score
    
    return pbp_df

# Example usage
chunked_pbp_scores = add_scores_to_pbp(chunked_pbp)

# print the first 5 and last 5 rows in the chunked play by play
display(chunked_pbp_scores.head())
display(chunked_pbp_scores.tail())
```
